# Activity Classification - DNN1 Training

This notebook trains our first Deep Neural Network on the physical activity dataset.


## Load Data and Prepare Training Set

```python
%reset -f

import importlib
import sys, os

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
import activity_functions
importlib.reload(activity_functions)
from activity_functions import *
```

```python
activtity = load_data()
```

```python
df_train, df_test = create_train_test(activtity, test_ratio=0.2)
print(df_train.shape)
print(df_test.shape)
```

```python
X_train, y_train, X_test, y_test = prepare_for_train(df_train, df_test)
```

## HyperTune

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
```

```python
tuner = kt.Hyperband(
    build_model,
    objective="val_accuracy",
    max_epochs=100,
    factor=3,
    directory="my_dir",
    project_name="hyperband_tuning"
)

batch_sizes = [32, 64, 128]
for batch in batch_sizes:
    tuner.search(X_train, y_train_encoded, epochs=100, validation_split=0.2, batch_size=batch)

best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

## Best Hyperparameters

```python
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print("Best hyperparameters:")
print(f"  learning_rate: {best_hps.get('learning_rate')}")
print(f"  optimizer: {best_hps.get('optimizer')}")
print(f"  dropout: {best_hps.get('dropout')}")
```

## Model Fit

```python
best_model = tuner.hypermodel.build(best_hps)
callback = tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=5, restore_best_weights=True)
history = best_model.fit(X_train, y_train_encoded, epochs=50, validation_split=0.2, callbacks=[callback])
```

## Model Evaluation

```python
y_test_hat = best_model.predict(X_test)
y_test_pred = np.argmax(y_test_hat, axis=1)
compute_scores(y_test, y_test_pred, verbose=True)
```
